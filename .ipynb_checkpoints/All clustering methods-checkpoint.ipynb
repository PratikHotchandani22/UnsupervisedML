```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

from sklearn.metrics import silhouette_score

from sklearn.cluster import Birch

from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import MeanShift
from sklearn.preprocessing import StandardScaler

from sklearn.cluster import SpectralClustering
from sklearn.cluster import AffinityPropagation
```

```python
df = pd.read_csv("Mall_Customers.csv")
```

```python
df
```

## Data Cleaning


```python
na_counts = df.isna().sum()
```

```python
na_counts
```

No null data

```python
# Remove duplicates based on all columns
df_unique = df.drop_duplicates()
```

```python
df_unique
```

```python
## Drop CustomerID variable, as it is redundant in our analysis

df = df_unique.drop(columns=['CustomerID'])
```

```python

# Perform one-hot encoding on the "Gender" column
df_encoded = pd.get_dummies(df, columns=['Gender'])
```

```python
df_encoded
```

```python
train_df, test_df = train_test_split(df_encoded, train_size=150, test_size=50, random_state=42)
```

### Model - Cluster building

### 1. K- Means


```python
# Perform k-means clustering for different values of K
wcss = []
k_values = range(1, 11)
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(train_df)
    wcss.append(kmeans.inertia_)

# Plot the WCSS for different values of K
plt.plot(k_values, wcss)
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS')
plt.title('Elbow Method')
plt.show()
```

```python
# Perform k-means clustering for different values of K
silhouette_scores = []
k_values = range(2, 11)  # Adjusted range to exclude K=1
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(train_df)
    labels = kmeans.labels_
    score = silhouette_score(train_df, labels)
    silhouette_scores.append(score)

# Plot the silhouette scores for different values of K
plt.plot(k_values, silhouette_scores)
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Different Values of K')
plt.show()

# Find the best value of K based on the highest silhouette score
best_k = k_values[silhouette_scores.index(max(silhouette_scores))]
print(f"The best value of K is: {best_k}")
```

### from the plot above, we can see that for cluster value K = 5 we have the highest silhouette score!!

```python
k = 5
kmeans = KMeans(n_clusters=k, random_state=42)
y_pred = kmeans.fit_predict(train_df)
```

```python
y_pred
```

```python
kmeans.cluster_centers_
```

```python
X = train_df
```

```python

# Get the cluster labels and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Plot the decision boundary
x_min, x_max = X.iloc[:, 0].min() - 1, X.iloc[:, 0].max() + 1
y_min, y_max = X.iloc[:, 1].min() - 1, X.iloc[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel(), np.zeros_like(xx.ravel()), np.ones_like(xx.ravel()), np.ones_like(xx.ravel())])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.8)
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=labels, edgecolors='k')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='o', s=200, color='red', label='Centroids')
plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
plt.title('K-means Clustering Decision Boundary')
plt.legend()
plt.show()
```

```python
# Perform agglomerative clustering
n_clusters = 5  # Number of clusters
agglomerative = AgglomerativeClustering(n_clusters=n_clusters)
labels = agglomerative.fit_predict(X)

# Visualize the clusters
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=labels, cmap='viridis')
plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
plt.title('Agglomerative Clustering')
plt.show()
```

```python

# Convert the DataFrame to a C-contiguous NumPy array
X_array = np.ascontiguousarray(X.values)

# Perform BIRCH clustering
n_clusters = 5  # Number of clusters
birch = Birch(n_clusters=n_clusters)
labels = birch.fit_predict(X_array)

# Visualize the clusters
plt.scatter(X['Age'], X['Annual Income (k$)'], c=labels, cmap='viridis')
plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
plt.title('BIRCH Clustering')
plt.show()
```

```python

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform Mean Shift clustering
mean_shift = MeanShift()
labels = mean_shift.fit_predict(X_scaled)

# Visualize the clusters
plt.scatter(X['Age'], X['Annual Income (k$)'], c=labels, cmap='viridis')
plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
plt.title('Mean Shift Clustering')
plt.show()
```

```python

# Perform Affinity Propagation clustering
affinity_propagation = AffinityPropagation()
labels = affinity_propagation.fit_predict(X_scaled)

# Visualize the clusters
plt.scatter(X['Age'], X['Annual Income (k$)'], c=labels, cmap='viridis')
plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
plt.title('Affinity Propagation Clustering')
plt.show()
```

```python

# Perform Spectral Clustering
spectral_clustering = SpectralClustering(n_clusters=3, random_state=42)
labels = spectral_clustering.fit_predict(X_scaled)

# Visualize the clusters
plt.scatter(X['Age'], X['Annual Income (k$)'], c=labels, cmap='viridis')
plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
plt.title('Spectral Clustering')
plt.show()
```
